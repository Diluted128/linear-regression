## Linear Regression Algorithm

### Learning data in one dimension 

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
from sympy.plotting import plot3d 
from sklearn import datasets 
```

#### Generate learning data 

```python
sample_size = 30
def fnc(x):
    return 10*x**4+0.5*x**2+0.5e4*np.random.rand(sample_size)

x_train = np.linspace(0,5,sample_size)
y_train = fnc(x_train) 

x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)

y_train = y_train / 1000
```

#### Plot generated data

```python
plt.plot(x_train, y_train, '*')
plt.grid()
plt.title('Data points')
```

#### Generate function of mean squered error

```python
a, b = smp.symbols('a b', real=True)
Fmse = 0
for number in range(len(x_train)):
    Fmse += 1/2*(y_train.item(number) - (x_train.item(number)*a + b))**2
```

#### Ploting 3d function of mean squered error

```python
plot3d(Fmse,show=True)
```

#### Calculate derivative of a and b parameter

```python
dfda = smp.diff(Fmse,a)
dfdb = smp.diff(Fmse,b)
```

```python
dfda
```

```python
dfdb
```

#### Gradient descent algorithm

```python
optimum_a = 0
optimum_b  = 0
steps = 30000
alpha = 0.00001

for step in range(steps):
    Sra = dfda.subs([(a, optimum_a),(b,optimum_b)]).evalf()
    Srb = dfdb.subs([(a, optimum_a),(b,optimum_b)]).evalf()
    optimum_a -= Sra * alpha
    optimum_b -= Srb * alpha  

print(optimum_a)
print(optimum_b)
```

#### Plot results by linear function

```python
a, b, x = smp.symbols('a b x', real=True)
linear_function = a*x + b;
function_f = smp.lambdify((x,a,b), linear_function)
e = np.linspace(0,5,100)
w = function_f(e, a = optimum_a, b = optimum_b)
plt.plot(e,w)
plt.plot(x_train, y_train, '*')
plt.grid()
```

#### Variance score

```python
m = Fmse.subs([(a, optimum_a),(b, optimum_b)]).evalf()
print('Variance score:', m)
```

### Model for x in many dimensions

```python
class LinearRegression:
    
    def __init__(self, data, target):
        self.__coefficients = np.empty(1)
        self.__data = data
        self.__target = target
        self.__extend_by_ones()
    
    def __extend_by_ones(self):
        columnOfOnes = np.ones((self.__data.shape[0], 1))
        self.__data = np.append(self.__data, columnOfOnes, axis=1)
        
    def fit(self, alpha, iterations):
        
        self.__coefficients = np.zeros(self.__data.shape[1]) 
            
        tags = np.zeros(self.__data.shape[1])
        
        for inter in range(iterations):
            tags = np.zeros(self.__data.shape[1])
            for col in range(self.__data.shape[0]):
                for row in range(self.__data.shape[1]):   
                    tags[row] += ((self.__coefficients*self.__data[col]).sum() - self.__target[col])*self.__data[col][row]
            self.__coefficients -= (tags * alpha);
          
        
    def coef_(self):
        if self.__coefficients.size == 0:
            return "NaN"
        else:
            return self.__coefficients[:-1]
    
    
    def intercept_(self):
        if self.__coefficients.size == 0:
            return "NaN"
        else:
            return self.__coefficients[-1]
       
    
    def predict(self, x):
        if self.__coefficients.size == 0:
            return "NaN"
        else:
            prediction = self.__coefficients[10] ## intercept
            for i in range(len(self.__coefficients) - 1):
                prediction += (self.__coefficients[i] * x[i])
            return prediction    
    
    
    def r2_score(self):
        if self.__coefficients.size == 0:
            return "NaN"
        else:
            mse = 0;
            for col in range(self.__data.shape[0]):
                mse += ((self.__coefficients*self.__data[col]).sum() - self.__target[col])**2;
            return mse;        
        
```

#### Import patients dataset

```python
patients = datasets.load_diabetes()
```

```python
print(patients['data'].shape)
print(patients['target'].shape)
```

####  Train model

```python
lin_reg1 = LinearRegression(patients.data, patients.target);
lin_reg1.fit(0.0001, 100);

lin_reg2 = LinearRegression(patients.data, patients.target);
lin_reg2.fit(0.0001, 500);

lin_reg3 = LinearRegression(patients.data, patients.target);
lin_reg3.fit(0.0001, 5000);

print("Coefficients of trained model:", lin_reg1.coef_())
print("Intercept of trained model: ", lin_reg1.intercept_())    
print("Coefficients of trained model:", lin_reg2.coef_())
print("Intercept of trained model: ", lin_reg2.intercept_()) 
print("Coefficients of trained model:", lin_reg3.coef_())
print("Intercept of trained model: ", lin_reg3.intercept_()) 
```

 #### Test

```python
print("Variance score of first generated model:", lin_reg1.r2_score())
print("Variance score of second generated model:", lin_reg2.r2_score())
print("Variance score of third generated model:", lin_reg3.r2_score())
```
